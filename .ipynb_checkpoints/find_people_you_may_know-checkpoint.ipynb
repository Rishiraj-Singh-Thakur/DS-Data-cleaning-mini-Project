```python
import json
def load_data(filename):
    with open(filename,"r")as file:
        return json.load(file)

def find_people_you_may_know(user_id, data):
    user_friends = {}
    for user in data["users"]:
        user_friends[user["id"]] = set(user["friends"])

    if user_id not in user_friends:
        return []

    direct_friends = user_friends[user_id]
    sujjetions = {}

    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if mutual != user_id and mutual not in direct_friends:
                sujjetions[mutual] = sujjetions.get(mutual,0)+1

    sorted_sujj = sorted(sujjetions.items(),key=lambda x :x[1], reverse =True)

    return [user_id for user_id, _ in sorted_sujj]

data = load_data("codebook_data.json")
user_id = 4
recomm = find_people_you_may_know(user_id,data)
print(f"sujjetion{recomm}")
```
